json 파일들 형식 맞추는 코드

```python
import json

src_file = "ict_terms_2.json"
dst_file = "restored_glossary.json"
```

```python
with open(src_file, encoding="utf-8") as f:
    src = json.load(f)

items = src if isinstance(src, list) else src.get("glossary", [])

converted = [
    {
        "question": str(i.get("question","")).strip(),
        "answer": str(i.get("answer","")).strip()
    }
    for i in items
    if "question" in i and "answer" in i
]
```

```python
with open(dst_file, encoding="utf-8") as f:
    restored = json.load(f)
```

```python
restored += converted
```

```python
with open(dst_file, "w", encoding="utf-8") as f:
    json.dump(restored, f, ensure_ascii=False, indent=4)
```
