```python
input_file = "finetune_glossary.jsonl"
```

```python
data = []
with open(input_file, "r", encoding="utf-8") as f:
    for line in f:
        line = line.strip()
        if line:
            data.append(json.loads(line))
```

```python
pattern = re.compile(r"^([^\s(]+)")

term_map = defaultdict(list)

for idx, item in enumerate(data):
    q = item.get("question", "").strip()

    m = pattern.match(q)
    if m:
        term = m.group(1)
    else:
        term = q

    term = term.split("(")[0].strip()

    term_map[term].append((idx, q))
```

```python
print("중복된 '첫 단어' 용어:")
for term, occ in term_map.items():
    if len(occ) > 1:
        print(f"\n[용어] {term}  ({len(occ)}개)")
        for idx, q in occ:
            print(f"  - index {idx}: {q}")
```
