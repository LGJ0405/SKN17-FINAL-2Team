```python
import json
import random

# 파일 경로 지정
input_file_path = "keep_TTA표준_ㅈ.jsonl"
output_file_path = "cleaned_keep_TTA표준_ㅈ.jsonl"  # 새로운 파일 경로 지정

# 중복을 확인할 딕셔너리
seen = set()
duplicates = []

# jsonl 파일 열기
with open(input_file_path, 'r', encoding='utf-8') as f:
    for line in f:
        try:
            # 한 줄씩 JSON 객체로 로드
            data = json.loads(line)
            
            # 'content' 값을 기준으로 중복 확인
            content = data['messages'][0]['content']  # 'content'는 1번 인덱스에 위치
            
            if content in seen:
                duplicates.append(data)  # 중복된 항목 추가
            else:
                seen.add(content)
        except json.JSONDecodeError:
            continue

# 중복된 항목에서 랜덤으로 하나만 남기고 나머지 제거하는 반복문
remaining_duplicates = []

while duplicates:
    print(f"중복된 항목 {len(duplicates)}개가 발견되었습니다. 랜덤으로 하나만 남깁니다.")
    
    # 중복 항목 중에서 랜덤으로 하나를 선택
    random_duplicate = random.choice(duplicates)

    # 결과 출력 (남은 항목)
    print("남은 항목:")
    print(random_duplicate)

    # 남은 항목만 두고 나머지 제거
    remaining_duplicates.append(random_duplicate)
    duplicates = [dup for dup in duplicates if dup != random_duplicate]

    # 중복이 없으면 반복문 종료
    if not duplicates:
        print("중복된 항목이 없습니다.")
        break

```

```python
# 중복 제거 후 데이터를 새로운 jsonl 파일에 저장
with open(output_file_path, 'w', encoding='utf-8') as out_f:
    for item in remaining_duplicates:
        json.dump(item, out_f, ensure_ascii=False)
        out_f.write('\n')

print(f"중복 항목을 제거한 데이터를 {output_file_path} 파일로 저장했습니다.")
```

```python
import json
import random

# 파일 경로 지정
input_file_path = "keep_TTA표준_ㅈ.jsonl"
output_file_path = "cleaned_keep_TTA표준_ㅈ.jsonl"  # 새로운 파일 경로 지정

# 중복을 확인할 딕셔너리
seen = set()
cleaned_data = []

# 중복을 저장할 리스트
duplicates = []

# jsonl 파일 열기
with open(input_file_path, 'r', encoding='utf-8') as f:
    for line in f:
        try:
            # 한 줄씩 JSON 객체로 로드
            data = json.loads(line)
            
            # 'content' 값을 기준으로 중복 확인
            content = data['messages'][0]['content'].replace(" ", "")  # 'content'는 1번 인덱스에 위치
            
            if content in seen:
                duplicates.append(data)  # 중복된 항목 저장
            else:
                seen.add(content)
                cleaned_data.append(data)  # 중복이 아닌 항목 저장
        except json.JSONDecodeError:
            continue

# 중복된 항목에서 랜덤으로 하나만 남기고 나머지 제거
remaining_duplicates = []

while duplicates:
    print(f"중복된 항목 {len(duplicates)}개가 발견되었습니다. 랜덤으로 하나만 남깁니다.")
    
    # 중복 항목 중에서 랜덤으로 하나를 선택
    random_duplicate = random.choice(duplicates)

    # 남은 항목만 두고 나머지 제거
    remaining_duplicates.append(random_duplicate)
    duplicates = [dup for dup in duplicates if dup != random_duplicate]

    # 중복이 없으면 반복문 종료
    if not duplicates:
        print("중복된 항목이 없습니다.")
        break

# 최종적으로 중복을 하나만 남긴 데이터를 포함한 리스트를 새로운 jsonl 파일에 저장
with open(output_file_path, 'w', encoding='utf-8') as out_f:
    # 원본 데이터를 기반으로, 중복을 하나만 남긴 리스트를 추가하여 저장
    for item in cleaned_data:
        json.dump(item, out_f, ensure_ascii=False)
        out_f.write('\n')

    # 중복된 항목들 중 하나를 남긴 데이터도 추가
    for item in remaining_duplicates:
        json.dump(item, out_f, ensure_ascii=False)
        out_f.write('\n')

print(f"중복 항목을 제거한 데이터를 {output_file_path} 파일로 저장했습니다.")
```
