```python
import json
```

```python
file_path = "__keep_TTA표준_ㅈ.jsonl"
```

```python
# 중복을 확인할 딕셔너리
seen = set()
duplicates = []
tete = []

# jsonl 파일 열기
with open(file_path, 'r', encoding='utf-8') as f:
    for line in f:
        try:
            # 한 줄씩 JSON 객체로 로드
            data = json.loads(line)
            
            # 'content' 값을 기준으로 중복 확인
            content = data['messages'][0]['content'].replace(" ", "")  # 'content'는 0번 인덱스에 위치
            
            if content in seen:
                duplicates.append(data)
            else:
                seen.add(content)
                tete.append(data)
        except json.JSONDecodeError:
            continue

# 중복된 항목 출력
if duplicates:
    print(f"중복된 항목 {len(duplicates)}개:")
    for dup in duplicates:
        print(dup['messages'][0]['content'])
else:
    print("중복된 항목이 없습니다.")
```
